# Array-based Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedListArray:
    def __init__(self,size):
        self.first = None
        self.next_free = 0
        self.array = []
        self.count = 0
        for i in range(size):
            new_node = Node(None)
            new_node.next = i+1 if i<size-1 else None
            self.array.append(new_node)

    def is_empty(self):
        return self.first == None
    
    def is_full(self):
        return self.next_free is None
    
    def insert_first(self, node):
        if self.is_full():
            return

        tmp = self.next_free
        self.next_free = self.array[tmp].next
        self.array[tmp] = node
        self.array[tmp].next = self.first
        self.first = tmp

        self.count += 1
        
    def insert_last(self, node):
        if self.is_full():
            return
    
        tmp = self.next_free
        self.next_free = self.array[tmp].next
        self.array[tmp] = node
    
        if self.is_empty():
            self.first = tmp
        else:
            curr = self.first
            while self.array[curr].next is not None:
                curr = self.array[curr].next
            self.array[curr].next = tmp
    
        self.count += 1
         
    
    def delete(self, data):
        if self.is_empty():
            return
    
        if self.array[self.first].data == data:
            tmp = self.first
            self.first = self.array[tmp].next
    
            self.array[tmp].next = self.next_free
            self.next_free = tmp
    
            self.count -= 1
            return
    
        curr = self.first
    
        while curr is not None and self.array[curr].next is not None:
            nxt = self.array[curr].next
    
            if self.array[nxt].data == data:
                self.array[curr].next = self.array[nxt].next
    
                self.array[nxt].next = self.next_free
                self.next_free = nxt
    
                self.count -= 1
                return
    
            curr = nxt
        
    def find(self, data):
        curr = self.first
    
        while curr is not None:
            if self.array[curr].data == data:
                return curr
            curr = self.array[curr].next
    
        return None
    
    def display(self):
        if self.is_empty():
            return
    
        lst = []
        curr = self.first
    
        while curr is not None:
            lst.append(self.array[curr].data)
            curr = self.array[curr].next
    
        return "\n".join(map(str, lst))
```

# Sorted Array-based Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedListArraySorted:
    def __init__(self,size):
        self.first = None
        self.next_free = 0
        self.array = []
        self.count = 0
        for i in range(size):
            new_node = Node(None)
            new_node.next = i+1 if i<size-1 else None
            self.array.append(new_node)

    def is_empty(self):
        return self.first == None
    
    def is_full(self):
        return self.next_free is None
    
    def insert(self, node):
        if self.is_full():
            return
    
        tmp = self.next_free
        self.next_free = self.array[tmp].next
        self.array[tmp] = node
    
        if self.is_empty() or node.data < self.array[self.first].data:
            self.array[tmp].next = self.first
            self.first = tmp
    
        else:
            prev = self.first
            curr = self.array[self.first].next
    
            while curr is not None and self.array[curr].data < node.data:
                prev = curr
                curr = self.array[curr].next
    
            self.array[tmp].next = curr
            self.array[prev].next = tmp
    
        self.count += 1
         
    
    def delete(self, data):
        if self.is_empty():
            return
    
        if self.array[self.first].data == data:
            tmp = self.first
            self.first = self.array[tmp].next
    
            self.array[tmp].next = self.next_free
            self.next_free = tmp
    
            self.count -= 1
            return
    
        curr = self.first
    
        while curr is not None and self.array[curr].next is not None:
            nxt = self.array[curr].next
    
            if self.array[nxt].data == data:
                self.array[curr].next = self.array[nxt].next
    
                self.array[nxt].next = self.next_free
                self.next_free = nxt
    
                self.count -= 1
                return
    
            curr = nxt   

    def find(self, data):
        curr = self.first
    
        while curr is not None:
            if self.array[curr].data == data:
                return curr
            curr = self.array[curr].next
    
        return None
    
    def display(self):
        if self.is_empty():
            return
    
        lst = []
        curr = self.first
    
        while curr is not None:
            lst.append(self.array[curr].data)
            curr = self.array[curr].next
    
        return "\n".join(map(str, lst))
```

# Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.first = None

    def is_empty(self):
        return self.first is None
    
    def insert_first(self,node):
        node.next = self.first
        self.first = node
        return
    
    def insert_last(self,node):
        if self.is_empty():
            self.first = node
        else:
            curr = self.first
            while curr is not None:
                curr = curr.next
            curr.next = node
    
    def delete(self, data):
        if self.is_empty():
            return
    
        if self.first.data == data:
            self.first = self.first.next
            return
    
        prev = self.first
        curr = self.first.next
    
        while curr is not None:
            if curr.data == data:
                prev.next = curr.next
                return
    
            prev = curr
            curr = curr.next

    def find(self,data):
        if self.is_empty():
            return
        curr = self.first
        while curr is not None:
            if curr.data == data:
                return curr
            curr = curr.next
    
    def display(self):
        curr = self.first
        lst = []
        while curr is not None:
            lst.append(curr.data)
            curr = curr.next
        return "\n".join(map(str,lst))
```

# Sorted Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedListSorted:
    def __init__(self):
        self.first = None

    def is_empty(self):
        return self.first is None
       
    def insert(self, node):
        if self.is_empty() or node.data < self.first.data:
            node.next = self.first
            self.first = node
            return
    
        prev = self.first
        curr = self.first.next
    
        while curr is not None and curr.data < node.data:
            prev = curr
            curr = curr.next
    
        node.next = curr
        prev.next = node

    def delete(self, data):
        if self.is_empty():
            return
    
        if self.first.data == data:
            self.first = self.first.next
            return
    
        prev = self.first
        curr = self.first.next
    
        while curr is not None:
            if curr.data == data:
                prev.next = curr.next
                return
    
            prev = curr
            curr = curr.next

    def find(self,data):
        if self.is_empty():
            return
        curr = self.first
        while curr is not None:
            if curr.data == data:
                return curr
            curr = curr.next
    
    def display(self):
        curr = self.first
        lst = []
        while curr is not None:
            lst.append(curr.data)
            curr = curr.next
        return "\n".join(map(str,lst))
    
    def insert(self,node):
        if self.is_empty() or self.first.data > node.data:
            node.next = self.first
            self.first = node
        else:
            curr = self.first
            while curr.next is not None and curr.next.data < node.data:
                curr = curr.next
            node.next = curr.next
            curr.next = node
```

# Single Class Linked List

```python
class SClinkedlistNode:
    def __init__(self, data=None):
        self.data = data
        self.next = None

    # create an empty list (sentinel)
    @classmethod
    def create_list(cls):
        return cls()  # sentinel node

    def insert(self, data):
        curr = self
        while curr.next:
            curr = curr.next
        curr.next = SClinkedlistNode(data)

    def delete(self, data):
        prev = self
        curr = self.next

        while curr:
            if curr.data == data:
                prev.next = curr.next
                return
            prev, curr = curr, curr.next

    def display(self):
        curr = self.next
        out = []
        while curr:
            out.append(curr.data)
            curr = curr.next
        return "\n".join(map(str, out))
```
