```python
# Stack
class Stack:
    def __init__(self,size):
        self.array = [None]*size
        self.tail = 0
        self.size = size

    def is_empty(self):
        return self.tail == 0
    
    def is_full(self):
        return self.tail == self.size
    
    def push(self,val):
        if self.is_full():
            return False
        self.array[self.tail] = val
        self.tail += 1
        return True
    
    def pop(self):
        if self.is_empty():
            return False
        self.tail -= 1
        tmp = self.array[self.tail]
        self.array[self.tail] = None
        return tmp
    
    def peek(self):
        if self.is_empty():
            return False
        return self.array[self.tail - 1]
    
    def display(self):
        return self.array


```

```python
# Queue
class Queue:
    def __init__(self,size):
        self.array = [None]*size
        self.head = 0
        self.tail = 0
        self.size = size

    def is_empty(self):
        return self.head == self.tail
    
    def is_full(self):
        return self.tail - self.head == self.size
    
    def enqueue(self,val):
        if self.is_full():
            return False
        self.array[self.tail] = val
        self.tail += 1
        return True
    
    def dequeue(self):
        if self.is_empty():
            return False
        tmp = self.array[self.tail]
        self.array[self.tail] = None
        self.tail -= 1
        return tmp
    
    def display(self):
        print(self.array)
```

```python
# Queue - Stack
class StackQueue:
    def __init__(self,size):
        self.enqueueStack = Stack(size)
        self.dequeueStack = Stack(size)
        
    def is_empty(self):
        return self.enqueueStack.is_empty()
    
    def is_full(self):
        return self.enqueueStack.is_full()
    
    def enqueue(self,val):
        if self.is_full():
            return False
        self.enqueueStack.push(val)
        return True

    def dequue(self):
        if self.is_empty():
            return False
        while not self.enqueueStack.is_empty():
            tmp = self.enqueueStack.pop()
            self.dequeueStack.push(tmp)
        tmp = self.dequeueStack.pop()
        while not self.dequeueStack.is_empty():
            tmp = self.dequeueStack.pop()
            self.enqueueStack.push(tmp)
        return tmp
    
    def display(self):
        self.enqueueStack.display()
```

```python
# Linked List
class LLNode:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.root = None
    
    def is_empty(self):
        return self.root is None
    
    def insert_first(self,data):
        if self.is_empty:
            self.root = LLNode(data)
            return True
        tmp = LLNode(data)
        tmp.next = self.root
        self.root = tmp
        return True

    def insert_last(self,data):
        if self.is_empty():
            self.root = LLNode(data)
            return True
        curr = self.root
        while curr.next is not None:
            curr = curr.next
        curr.next = LLNode(data)
        return True
    
    def display(self):
        if self.is_empty():
            return False
        result = []
        curr = self.root
        while curr.next is not None:
            result.append(curr)
        return result
        
```
