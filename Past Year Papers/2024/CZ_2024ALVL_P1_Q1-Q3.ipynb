**2 (c)** <br>
When S(2) is called, it will add itself to the function call stack, then subsequently call S(1), adding that to the call stack. Since n now is equal to 1, S(1) will return 12, which results in the function call getting popping from the call stack. Then S(2) will now return 12 - 3 = 9 as shown in the trace tree, as S(2-1) is resolved as 9.

**2 (d)**<br>
When S(0) is called, it will recurse indefinetely and may eventually reach the recursion limit if one exists. This is because, after S(0) is called, since n is not equal to 1, S(0 - 1) - 3 is returned, which subsequently returns S(-1-1) - 3. Thus, n will become negatively and the terminating case of n equal 0 will never be reached.
