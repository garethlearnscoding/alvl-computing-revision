A denary number can be converted to binary using the division by 2 method. The denary number is divided by 2 repeatedly until it becomes 0. After each division, the remainder of 1 or 0 is stored. These remainders are written in reverse order to create the binary number.

For example, converting the denary number 28 to binary:

```text
28/2=14 remainder 0
14/2=7 remainder 0
7/2=3 remainder 1
3/2=1 remainder 1
1/2=0 remainder 1
28 denary is 11100 in binary.
```
