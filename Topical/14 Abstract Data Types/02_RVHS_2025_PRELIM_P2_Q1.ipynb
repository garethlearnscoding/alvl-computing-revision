```python
class Fish:
    def __init__(self,name,size):
        self.name = name
        self.size = size
    def get_name(self):
        return self.name
    def get_size(self):
        return self.size
    def consume(self,fish):
        if self.size > fish.get_size():
            return Fish(self.name,self.size+fish.get_size())
        else:
            return Fish(fish.get_name,self.size+fish.get_size())
    def __str__(self):
        return f"Name:{self.get_name():<35} Size: {self.get_size():>1}" 
```
