```python
class Fish:
    def __init__(self,name,size):
        self.name = name
        self.size = int(size)
    def name_getter(self):
        return self.name
    def size_getter(self):
        return self.size
    def __str__(self):
        return f"Name:{self.name:<20}Size:{self.size:>4}"
    def consume(self,fish):
        othername = fish.name_getter()
        othersize = fish.size_getter()
        if othersize > self.size:
            return Fish(othername,int(self.size)+ int(othersize))
        else:
            return Fish(self.name,int(self.size)+int(othersize))
    
```

 * display the information of this last fish using `str(fish)`
 * Call `task1_3()` 10 times.

```python
import random
import csv

class Node:
    def __init__(self,data):
        self.data = data
        self.next_ptr = -1

class SortedLinkedListForFish:
    def __init__(self):
        self.start = -1
        self.next_free = 0
        self.count = 0
        self.nodes = [Node(None)]
        for i in range(49):
            self.nodes[i].next_ptr = i+1 if i < 49 else -1
            new = Node(None)
            self.nodes.append(new)

    def is_empty(self):
        return self.start == -1
    
    def is_full(self):
        return self.next_free == -1
    
    def insert(self,fish):
        if self.is_full():
            pass
        elif self.is_empty() or self.nodes[self.start].data.size_getter() > fish.size_getter():
            tmp = self.next_free
            self.next_free = self.nodes[tmp].next_ptr
            self.nodes[tmp] = Node(fish)
            self.nodes[tmp].next_ptr = self.start
            self.start = tmp
            self.count += 1
        else:
            prev = self.start
            curr = self.nodes[self.start].next_ptr
            while curr != -1 and fish.size_getter() > self.nodes[curr].data.size_getter():
                prev = curr
                curr = self.nodes[curr].next_ptr
            tmp = self.next_free
            self.next_free = self.nodes[tmp].next_ptr
            self.nodes[tmp] = Node(fish)
            self.nodes[tmp].next_ptr = curr
            self.nodes[prev].next_ptr = tmp
            self.count += 1
        if self.count == 50:
            self.next_free = -1

    def delete(self,fish_name):
        if self.is_empty():
            return
        elif self.nodes[self.start].data.name_getter() == fish_name:
            tmp = self.start
            self.start = self.nodes[tmp].next_ptr
            self.nodes[tmp].next_ptr = self.next_free
            self.next_free = tmp
            self.count -= 1
        else:
            prev = self.start
            curr = self.nodes[self.start].next_ptr
            while curr != -1 and self.nodes[curr].data.name_getter() != fish_name:
                prev = curr
                curr = self.nodes[curr].next_ptr
            if curr == -1:
                return
            else:
                self.nodes[prev].next_ptr = self.nodes[curr].next_ptr
                self.nodes[curr].next_ptr = self.next_free
                self.next_free = curr
                self.count -= 1
    
            
    def display(self):
        result = []
        curr = self.start
        while curr != -1:
            result.append(str(self.nodes[curr].data))
            curr = self.nodes[curr].next_ptr
        return "\n".join(result)
        
    def pick2RandomFish(self):
        all_fishes = [i for i in self.nodes if i.data is not None]
        samply = random.sample(all_fishes,2)
        randfishes = [i.data for i in samply]
        return randfishes
    
    def preyTime(self):
        if self.count == 1:
            return True
        randfishes = self.pick2RandomFish()
        # print([type(i.size_getter()) for i in randfishes])
        for fish in randfishes:
            self.delete(fish.name_getter())
        f1,f2 = randfishes
        f3 = f1.consume(f2)
        self.insert(f3)
    
def task1_3():
    with open("./resources/fishes.txt") as f:
        data = [[i[0],int(i[1])] for i in csv.reader(f)]
    fishy = SortedLinkedListForFish()
    for fish in data:
        fishyyyy = Fish(*fish)
        fishy.insert(fishyyyy)
        # print(type(fishyyyy.size_getter()))
    while True:
        res = fishy.preyTime()
        if res:
            break
    print(fishy.display())

for i in range(10):   
    task1_3()

task1_3()



```
