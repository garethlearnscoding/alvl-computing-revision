Write program code to implement the client’s `menu()` procedure which allows the client to establish a connection to the server and communicate based on the message protocol given.

The Mastermind client menu options are as follow. Each option will create a message of its own.
```text
Master Mind Game Menu
1) Just start a new game
2) Make a guess code
3) Display history
4) Quit
```

The following bullet points defines the Mastermind game protocol for the client.
- message: `START`
 - description: The `START\n` message is sent from the client to the server when client chooses option 1).

 The server will reply to the client in a `START` message with the number of attempts the client has before the game starts. The format of the message is as follow:
 ```text
 START,<number of attempts left>\n
 ```

 For example, the client program should display the following when option 1 is chosen and `START,5\n` is received.
 ```text
 Client initiates connection to server.
 Client connected to server.
 Master Mind Game Menu
 1) Just start a new game
 2) Make a guess code
 3) Display history
 4) Quit
 Type an option 1
 New game starts begins with 5 attempts.
 ```
- message: `GUESS`
 - description: The `GUESS` message is sent from the client to the server when client chooses option 2). The format of the message is as follow:
 ```text
 GUESS,<guesscode>\n
 ```

 For example, if the client input `1122` as the guess code, the message sent to the server is as follow:
 ```text
 GUESS,1122\n
 ```

 When the server received a `GUESS` message, there 3 scenarios.
 - **Case 1**
 If the guess code is correct, the server replies with a `WON\n` message. Upon receiving the `WON\n` message, the Client program quits
 ```text
 Master Mind Game Menu
 1) Just start a new game
 2) Make a guess code
 3) Display history
 4) Quit
 Type an option:2
 Make a guess:6441
 You have WON. The code is 6441.
 ```
 - **Case 2**
 If the guess code is incorrect and there is no more attempt left, the server replies with a `LOSE,<answer code>\n` message. Upon receiving the `LOSE\n` message, the Client program quits. E.g. `LOSE,3432\n`
 ```text
 Master Mind Game Menu
 1) Just start a new game
 2) Make a guess code
 3) Display history
 4) Quit
 Type an option:2
 Make a guess:5566
 You have run out of attemps. The code is 3432.
 ```
 - **Case 3**
 If it is not case 1 and 2, the server replies a `GUESS` message with the guess result. The guess result is a string that is made of “B”, “W” or “-”.

 The format of the message the server sent back to client is as follow:
 ```text
 GUESS,<result>\n
 ```

 For example, if the guess code `1122` has only 1 colour been guessed correctly at the correct position, “B” is returned by server. Then, the message that is sent from the server to the 
 client is as follow: 
 ```text
 GUESS,B\n
 ```

 The client program should display the following when option 2 is chosen, `1122` is input and `GUESS,B\n` is received.
 ```text
 Master Mind Game Menu
 1) Just start a new game
 2) Make a guess code
 3) Display history
 4) Quit
 Type an option:2
 Make a guess:1122
 Guess Result: 8
 ```

 This is another possible result of case 3 where option 2 is chosen, `5436` is input and `GUESS,BWW\n` is received.
 ```text
 Master Mind Game Menu
 1) Just start a new game
 2) Make a guess code
 3) Display history
 4) Quit
 Type an option:2
 Make a guess: 5436
 Guess Result: BWW
 ```

- message: `HISTORY`
 - description: The `HISTORY\n` message is sent from the client to the server when the client chooses option 3).

 The server will reply to the client in a `HISTORY` message with all previous guess codes and their results in the message. The format of the message is as follow:
 ```text
 HISTORY,<no_attempts_left>,<1st guess code>,<1st result>,<2nd guess code>,<2nd result>,...,<last guess code>,<last result>\n
 ```
 
 For example, the client program should display the following when option 3 is chosen and `HISTORY,3,1122,B,2233,BW,2525,----\n` is received.
 ```text
 Master Mind Game Menu
 1) Just start a new game
 2) Make a guess code
 3) Display history
 4) Quit
 Type an option:3
 You have 3 attempts left.
 1 1122 B
 2 2233 BM
 3 2525 ----
 ```

- message: `QUIT`
 - description: The `QUIT\n` message is sent from the client to the server when client chooses option 4). Both the client and the server program quit
 ```text
 Master Mind Game Menu
 1) Just start a new game
 2) Make a guess code
 3) Display history
 4) Quit
 Type an option:4
 Game program ends.
 ```
